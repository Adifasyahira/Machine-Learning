```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

from sklearn.svm import SVC: Ini mengimpor class SVC dari modul svm di dalam Scikit-learn. SVC merupakan singkatan dari Support Vector Classification, yang merupakan algoritma SVM (Support Vector Machine) yang digunakan untuk pemodelan klasifikasi.

Standardisasi mengubah fitur-fitur numerik sehingga memiliki rata-rata nol dan varians satu, dengan tujuan membuat pemrosesan lebih stabil dan mempengaruhi hasil algoritma yang sensitif terhadap skala seperti SVM.

from sklearn.model_selection import GridSearchCV: Ini mengimpor GridSearchCV, sebuah fungsi dalam Scikit-learn yang digunakan untuk melakukan pencarian parameter secara sistematis melalui ruang parameter yang telah ditentukan. GridSearchCV digunakan untuk mencari parameter terbaik untuk model, dengan mencoba semua kombinasi parameter yang mungkin yang telah ditentukan dalam grid.

Data train digunakan untuk melatih model, sementara data test digunakan untuk menguji kinerja model yang telah dilatih.

Kode ini digunakan untuk membaca data melalui file excel lalu melihat data 5 tearatas.

```python
X = df.drop("Class_Ürgüp Sivrisi", axis=1)
y = df["Class_Ürgüp Sivrisi"]
```

Kode ini digunakan untuk melakukan pemisahan antara fitur (features) dan target dalam sebuah dataset.X = df.drop("Class_Ürgüp Sivrisi", axis=1): Di sini, Anda sedang membuat variabel X yang akan berisi fitur-fitur dari DataFrame df. Anda menggunakan metode drop() dari pandas DataFrame untuk menghapus kolom yang disebut "Class_Ürgüp Sivrisi" dari dataset df. Secara khusus, dengan axis=1, Anda memberi tahu pandas untuk menghapus sebuah kolom (bukan baris), dan kolom yang dihapus adalah "Class_Ürgüp Sivrisi". Variabel X akan berisi semua kolom dari DataFrame df kecuali kolom target yang dihapus.

y = df["Class_Ürgüp Sivrisi"]: Di sini, Anda sedang membuat variabel y yang akan berisi target atau label dari dataset df. Anda menetapkan kolom "Class_Ürgüp Sivrisi" dari DataFrame df ke variabel y. Dengan melakukan ini, Anda memisahkan kolom "Class_Ürgüp Sivrisi" dari DataFrame sebagai target atau label yang ingin diprediksi.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=47)
```

Kode tersebut menggunakan train_test_split dari Scikit-learn untuk membagi dataset menjadi subset train dan test yang akan digunakan dalam proses pelatihan dan evaluasi model machine learning.Dengan pemisahan ini, Anda memiliki X_train dan y_train untuk melatih model, dan X_test serta y_test untuk menguji performa model pada data yang tidak pernah dilihat sebelumnya selama proses pelatihan.

```python
model = SVC()
scaler = StandardScaler()
```

model = SVC(): Ini membuat objek model dari kelas SVC() yang disediakan oleh Scikit-learn. SVC merupakan singkatan dari Support Vector Classifier, yang merupakan implementasi dari algoritma Support Vector Machine (SVM) untuk tugas klasifikasi. Objek model yang dibuat di sini akan digunakan untuk menginisialisasi model SVM yang akan dilatih dengan data.

scaler = StandardScaler(): Ini membuat objek dari kelas StandardScaler() yang disediakan oleh Scikit-learn. StandardScaler digunakan untuk melakukan standardisasi pada fitur-fitur numerik dalam dataset. Proses standardisasi mengubah fitur-fitur numerik sehingga memiliki rata-rata nol dan varians satu, membuat distribusi data menjadi standar yang membantu algoritma machine learning yang sensitif terhadap skala untuk bekerja lebih baik.

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

Dengan melakukan standarisasi ini, variabel scaled_X_train akan berisi data train yang telah diubah skala sedemikian rupa sehingga memiliki rata-rata nol dan varians satu berdasarkan informasi dari data train. Sementara itu, variabel scaled_X_test akan berisi data test yang telah diubah skala berdasarkan informasi yang sama dari data train untuk konsistensi dalam pemrosesan data

```python
model.get_params().keys()
```

Jika memiliki model SVC() yang telah diinisialisasi sebelumnya dan kemudian Anda menggunakan model.get_params().keys(), hasilnya akan berupa daftar kunci atau nama parameter yang dapat diatur dalam model SVM. Dengan mengetahui parameter-parameter ini, Anda dapat melakukan penyetelan hyperparameter untuk mengoptimalkan kinerja model, seperti menggunakan GridSearchCV untuk mencari kombinasi parameter terbaik, atau melakukan tuning manual berdasarkan pemahaman domain.

```python
param_grid = {"C": [0,0.1,1,10,100,1000],
              "degree":[2,3,4,5],
             "gamma": ["scale","auto"],
             "kernel": ['linear', 'poly', 'rbf']
             }
```

param_grid adalah sebuah dictionary yang berisi kombinasi dari parameter-parameter yang akan dieksplorasi selama proses pencarian parameter terbaik menggunakan GridSearchCV. Setiap parameter memiliki daftar nilai-nilai yang berbeda yang akan dieksplorasi oleh GridSearchCV untuk menemukan kombinasi parameter terbaik yang menghasilkan performa model yang optimal.Dengan menggunakan GridSearchCV dan param_grid seperti ini, Anda melakukan pencarian yang sistematis untuk menemukan kombinasi parameter terbaik yang akan digunakan untuk model SVM, berdasarkan performa yang diukur menggunakan teknik cross-validation.

```python
grid = GridSearchCV(model,param_grid, cv=5, scoring="accuracy")
```

Dengan menggunakan GridSearchCV dengan model SVC, param_grid, dan parameter lain yang telah dijelaskan, Anda melakukan pencarian parameter terbaik untuk model SVM dengan skema cross-validation 5-fold, dengan tujuan untuk memilih kombinasi parameter yang memberikan hasil akurasi terbaik dalam memprediksi target pada dataset yang digunakan.Pada scoring="accuracy": Parameter scoring menentukan metrik evaluasi yang digunakan untuk mengevaluasi performa model. 

```python
grid = GridSearchCV(model,param_grid, cv=5, scoring="accuracy")
grid.fit(scaled_X_train, y_train)
```

GridSearchCV akan melakukan proses pencarian kombinasi parameter terbaik dari param_grid yang telah Anda tentukan sebelumnya. Ini dilakukan dengan menggunakan skema cross-validation dengan 5-fold (sesuai dengan nilai cv=5 yang telah ditentukan sebelumnya). GridSearchCV akan mengevaluasi performa model dengan berbagai kombinasi parameter tersebut pada data train (scaled_X_train dan y_train) untuk setiap fold yang dibuat dalam cross-validation.

Tujuan utama dari pemanggilan grid.fit() adalah untuk melatih model dengan kombinasi parameter yang berbeda dan menentukan kombinasi parameter terbaik yang menghasilkan performa yang optimal, seperti yang diukur oleh metrik evaluasi (dalam hal ini, metrik akurasi yang ditentukan oleh scoring="accuracy").

```python
grid.best_params_
```

best_params_ adalah atribut dari objek GridSearchCV yang memberikan informasi tentang kombinasi parameter terbaik yang ditemukan selama proses pencarian parameter.

Ketika menggunakan GridSearchCV untuk mencari kombinasi parameter terbaik, setelah proses pelatihan selesai, atribut .best_params_ dari objek GridSearchCV akan memberikan dictionary yang berisi kombinasi parameter terbaik yang ditemukan oleh pencarian.

```python
from sklearn.metrics import classification_report, confusion_matrix
```

Kedua fungsi ini berguna dalam mengevaluasi kinerja model klasifikasi. classification_report memberikan laporan rinci yang mencakup beberapa metrik evaluasi, sedangkan confusion_matrix memberikan representasi visual yang lebih langsung tentang seberapa baik model mengklasifikasikan setiap kelas. Dengan bantuan kedua fungsi ini, Anda dapat memahami performa model dan menentukan di mana model tersebut mungkin perlu peningkatan atau penyesuaian lebih lanjut.

```python
y_pred = grid.predict(scaled_X_test)
print(classification_report(y_test, y_pred))
```

laporan tersebut akan menyajikan metrik evaluasi seperti presisi (precision), recall, F1-score, dan jumlah sampel yang mendukung (support) untuk setiap kelas dalam data test. Dengan menggunakan laporan ini, Anda dapat memahami seberapa baik model Anda bekerja untuk setiap kelas dan dapat mengevaluasi performa model secara menyeluruh

```python
confusion_matrix(y_test, y_pred)
```

Fungsi confusion_matrix(y_test, y_pred) dari modul sklearn.metrics digunakan untuk menghasilkan matriks konfusi (confusion matrix) berdasarkan perbandingan antara nilai sebenarnya (y_test) dan nilai yang diprediksi (y_pred) oleh model pada data test. Matriks konfusi ini memberikan gambaran yang jelas tentang seberapa baik model melakukan klasifikasi untuk setiap kelas. Dengan informasi dari matriks konfusi, Anda dapat mengevaluasi di mana model cenderung salah dan di mana model bekerja dengan baik dalam memprediksi kelas-kelas tertentu.
